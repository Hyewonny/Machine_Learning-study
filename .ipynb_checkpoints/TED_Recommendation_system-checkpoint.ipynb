```python
# New Test 할 문서들의 URL을 크롤링
def new_teset_url():
    url = []
    mytitle=[]
    for i in range(1,2):
        urllst = base_url+"/talks?page="+str(i)
        soup = BeautifulSoup(requests.get(urllst).text, 'html.parser')
        my_title = soup.select('h4 > a')
        for title in my_title:
            subtitle = title.get('href')
            mytitle.append(subtitle)
    for i in mytitle:
        url.append(base_url+i)
    return url
```

```python
# To make Training set, figure's video URL crawler.
def figure_extract(lst):
    urllst = []
    bs_url = base_url+'/talks?sort=newest&q='
    namelst = [i.replace(" ", '+') for i in lst]
    for i in range(len(namelst)):
        test = BeautifulSoup(requests.get(bs_url+namelst[i]).text, 'html.parser')
        testdata = test.select('h4 > a')
        for j in range(len(testdata)):
            urllst.append(base_url + testdata[j].get('href'))
    return urllst
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import wordpunct_tokenize
# stemming 형태소 분석
from nltk.stem.porter import PorterStemmer
```

```python
# remove stopword and stemming processing.
def word2_vec(string):
    stop_words = set(stopwords.words('english'))
    stop_words.update(['ted', 'talks', 'talk','tedx'])
    sp_md = string.split()
    sp_md = [i for i in sp_md if i not in stop_words]
    porter = PorterStemmer() # define an object
    j=[]
    for doc in sp_md:
        j.append([porter.stem(i) for i in wordpunct_tokenize(doc)])
        vocabSet = set(list(np.unique(j)))# is that weird...ha?
    return vocabSet
```

```python
# making training set 
def mak_data(url):
    test = BeautifulSoup(requests.get(url).text, 'html.parser')
    theme_data = test.select('head > meta')
    speaker_url = test.select('span > link')
    meta_data = str(theme_data[1]).split("<meta content")[2:4]
    first = re.sub('[^a-z]+'," ",meta_data[0].split('"')[1].lower())
    secon = re.sub('[^a-z]+'," ",meta_data[1].split('"')[1].lower())
    url2 = str(speaker_url[0])
    url2 = base_url+re.search('\/\w*\W\w*', url2).group()
    test = BeautifulSoup(requests.get(url2).text, 'html.parser')
    spk_data = str(test.find("div", {"class":"profile-intro"}).text)
    spk_data = re.sub('[^a-z]+'," ",spk_data.lower())
    meta_data_li = first+' '+secon+' '+spk_data+' '
    vocabSet = word2_vec(meta_data_li)
    return(vocabSet)
```

```python
def trainNB0(trainMatrix, trainCategory):
    numTrainDocs = len(trainMatrix)
    numWords = len(trainMatrix[0])
    prior = sum(trainCategory)/float(numTrainDocs)
    # initialize to 1
    p0Num = np.ones(numWords);p1Num = np.ones(numWords)
    # value to add if it is in vector.
    p0Denom = 2.0;p1Denom = 2.0
    for i in range(numTrainDocs):
        if trainCategory[i] == 1:
            p1Num +=trainMatrix[i]
            p1Denom += sum(trainMatrix[i])
        else:
            p0Num += trainMatrix[i]
            p0Denom += sum(trainMatrix[i])
    p1Vect = np.log(p1Num/p1Denom)
    p0Vect = np.log(p0Num/p0Denom)
    return p0Vect, p1Vect, prior 
```

```python
def classifyNB(vec2Classify, p0Vec, p1Vec, pClass1):
    value_lst = []
    for i in range(len(vec2Classify)):
        p1 = sum(vec2Classify[i]*p1Vec) + np.log(pClass1)
        p0 = sum(vec2Classify[i]*p0Vec) + np.log(1-pClass1)
        if p1>p0: value = 1
        else: value = 0; 
        value_lst.append(value)
    return value_lst
```

```python
def classifyNB0(vec2Classify, p0Vec, p1Vec):
    value_lst = []
    for i in range(len(vec2Classify)):
        p1 = sum(vec2Classify[i]*p1Vec)
        p0 = sum(vec2Classify[i]*p0Vec)
        if p1>p0: value = 1
        else: value = 0; 
        value_lst.append(value)
    return value_lst
```
